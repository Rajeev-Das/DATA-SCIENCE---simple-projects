## ANALYSIS OF THE IPL TEAMS

```python
# The Team with number of matches won. 
Teams_won = df_m['winner'].value_counts()
Teams_won
```

### We can see that Rising Pune Supergiants have been repeated twice and hence we need to edit that first.

```python
df_m['winner'].replace(to_replace='Rising Pune Supergiant', value='Rising Pune Supergiants', inplace=True)
Teams_won = df_m['winner'].value_counts()
Teams_won
```

```python
print('Number of teams participated: ', len(Teams_won))
```

```python
Teams_won = Teams_won.plot.barh(figsize=(15,8), color='crimson', fontsize=15)
plt.style.use('dark_background')
Teams_won.set_title('Teams with maximum win\'s', fontsize = 15)
Teams_won.set_xlabel('Number of matches won', fontsize=15)
Teams_won.invert_yaxis()
```

```python
# The team which won by maximum runs
df_m['win_by_runs'].max()
```

```python
df_m.iloc[df_m['win_by_runs'].idxmax()]
```

### The winner was Mumbai Indians for the match which was won by maximum runs. 

```python
# The team which won by maximum wickets
df_m['win_by_wickets'].max()
```

```python
df_m.iloc[df_m['win_by_wickets'].idxmax()]
```

### The team Kolkata Knight Riders won by the maximum wickets.

```python
# Checking for the more teams which won by 10 wickets
df_m[df_m.win_by_wickets==10]['winner'].value_counts()
# RCB won 3 times by 10 wickets. 
```

```python
# Teams which won the trophies
ipl = df_m[['season', 'winner']]
#creaing an empty dictionary
teamswon_ipl = {}
for i in sorted(ipl.season.unique()):
    teamswon_ipl[i] = ipl[ipl.season == i]['winner'].tail(1).values[0]
season_winner = pd.Series(teamswon_ipl)
season_winner = pd.DataFrame(season_winner, columns=['Season_winner'])
print(season_winner['Season_winner'].unique())
print('The number of teams which won a title: ', season_winner['Season_winner'].nunique())
season_winner
```

```python
# Finding out the number of toss wins w.r.t teams
df_m['toss_winner'].value_counts()
```

```python
# Toss decisions made
Toss_decision = df_m['toss_decision'].value_counts()
Toss_decision
```

```python
print('Percentage when teams choose to field: ', Toss_decision['field']/Toss_decision.sum())
print('Percentage when teams choose to bat: ', Toss_decision['bat']/Toss_decision.sum())
Toss_decision = Toss_decision.plot.barh(figsize=(12,5), color='orchid', fontsize=15)
plt.style.use('dark_background')
Toss_decision.set_title('Toss decision made', fontsize = 15)
Toss_decision.set_ylabel('% Count of Toss Selection', fontsize=15)
#Toss_decision.invert_yaxis()
```
